```python
import seaborn as sns
import numpy as np
val_size = np.array([35, 56, 120, 12, 122, 737, 15])
test_size = np.array([43, 93, 217, 44, 171, 908, 35])
train_sm_size = np.array([292, 458, 979, 103, 991, 5968, 127])
train_aug_size = np.array([7812, 7986, 8012, 6898, 7948, 8008, 7279])

sns.set_style("whitegrid")  # Use grid style
```

```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.DataFrame({
    'Validation Set': val_size,
    'Test Set': test_size,
    'Training Set':train_sm_size,
    'Training Set After Augmentation':train_aug_size,
    'category': ['akiec', 'bcc', 'bkl', 'df', 'mel', 'nv', 'vasc']
})
```

```python
long_df = pd.melt(df, id_vars=['category'], 
                  value_vars=['Validation Set', 'Test Set', 'Training Set'],
                  var_name='dataset', value_name='size')

# Create stacked histogram
sns.barplot(data=long_df, x='category', y='size', hue='dataset')
plt.xlabel("Lesion Category")
plt.ylabel("Number of Images")
plt.title("Stacked Histogram of Dataset Sizes per Category")
plt.show()
```

```python
sns.barplot(data=df, x='category', y='Training Set After Augmentation')
plt.xlabel("Leision Category")
plt.ylabel("Number of Images")
plt.title("Augmented Training Set Size per Category")
plt.show()
```
